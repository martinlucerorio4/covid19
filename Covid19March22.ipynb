```python
import numpy as np
import pandas as pd


covid=pd.read_csv(r'C:\Users\andre\Downloads\novel-corona-virus-2019-dataset\time_series_covid_19_confirmed.csv')
cases=covid[['Country/Region','Lat','Long','3/22/20']]
cases=cases.rename(columns={'3/22/20':'Cases'})
cases.head()
```

```python
!pip install folium
import folium
from folium import features
```

```python
cases_map=folium.Map(zoom_start=12)
cases_map
```

```python
cases2=folium.map.FeatureGroup()
```

```python
for lat,lng in zip(cases.Lat, cases.Long):
    cases2.add_child(
        folium.CircleMarker(
            [lat,lng],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6,
        )
    )
    

latitudes=list(cases.Lat)
longitudes=list(cases.Long)
labels=list(cases.Cases)

for lat,lng, label in zip(latitudes,longitudes,labels):
    folium.Marker([lat,lng],Cases=label).add_to(cases_map)

cases_map.add_child(cases2)
```

```python
cases2_map=folium.Map(zoom_start=12)

for lat, lng, label in zip(cases.Lat,cases.Long, cases.Cases):
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    color='yellow',
    fill=True,
    popup=label,
    fill_color='blue',
    fill_opacity=0.6,
    ).add_to(cases_map)
    
cases_map
```

```python
from folium import plugins


cases_map=folium.Map(zoom_start=20)
cases1=plugins.MarkerCluster().add_to(cases_map)

for lat,lng,label, in zip(cases.Lat, cases.Long,cases.Cases):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        pupup=label,
    ).add_to(cases1)
    
cases_map
```
